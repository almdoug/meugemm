```python
# Importar bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Função para ler e processar arquivos .dat
def processar_dados(arquivo, diretorio):
    df = pd.read_csv(f'{diretorio}/{arquivo}', sep=',', engine='python')
    matSize = df['matSize'].astype(int)
    gflops = df['Size'].astype(float)
    tempo_medio = df['Mean'].astype(float)
    tempo_max = df['Largest'].astype(float)
    tempo_min = df['Smallest'].astype(float)
    
    # Desempenho = GFLOPS / Tempo (GFLOPS/s)
    desempenho_medio = gflops / tempo_medio
    desempenho_min = gflops / tempo_max
    desempenho_max = gflops / tempo_min
    
    return matSize, desempenho_medio, desempenho_min, desempenho_max
```

## Comparação: Nativo vs Docker - Alternatives

```python
# Carregar dados - ATLAS (Alternatives)
matSize_atlas_nat, desemp_atlas_nat, _, _ = processar_dados('output_ATLAS.dat', '../output/native/alternatives')
matSize_atlas_dock, desemp_atlas_dock, _, _ = processar_dados('output_ATLAS.dat', '../output/docker/alternatives')

plt.figure(figsize=(10,6))
plt.plot(matSize_atlas_nat, desemp_atlas_nat, '-', label='Nativo', marker='o')
plt.plot(matSize_atlas_dock, desemp_atlas_dock, '--', label='Docker', marker='s')
plt.xlabel('Tamanho da matriz')
plt.ylabel('Desempenho (GFLOPS/s)')
plt.title('ATLAS - Alternatives: Nativo vs Docker')
plt.legend()
plt.grid(True)
plt.show()
```
