```python
# Importar bibliotecas necess√°rias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Ler o arquivo BLAS.dat como CSV
df = pd.read_csv('BLAS.dat', sep=',', engine='python')
df.head()
```

```python
size = df['matSize'].astype(int)
flop = df['Size'].astype(float)
timeBlas = df['Mean'].astype(float)
timeMax = df['Largest'].astype(float)
timeMin = df['Smallest'].astype(float)
flopsBlas = flop/timeBlas
flopsMin = flop/timeMax
flopsMax = flop/timeMin
```

```python
# Plotar gr√°fico da m√©dia, menor e maior tempo de execu√ß√£o por matSize
plt.figure(figsize=(10,6))
plt.plot(size, flopsBlas, label='M√©dia', marker='o')
plt.plot(size, flopsMin, '--', label='Menor tempo')
plt.plot(size, flopsMax, '--', label='Maior tempo')
plt.xlabel('Tamanho da matriz (matSize)')
plt.ylabel('Desempenho (GFLOPS)')
plt.title('Evolu√ß√£o da m√©dia, menor e maior tempo de execu√ß√£o - DGEMM Blas')
plt.legend()
plt.grid(True)
plt.show()
```
